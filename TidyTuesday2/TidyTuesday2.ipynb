As a Kdrama lover myself, the question to be answered for this notebook is "How many South Korean shows and movies were added each year? Secondly, generally when were the added movies and shows initially released?"

<h3>PYTHON LIBRARIES</h3>

```python
# Data analysis and data preparation
import pandas as pd
import numpy as np
import random as rnd #for machine learning

# Data visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3>DATA ANALYSIS I</h3>

```python
#show last 10 rows of data

netflix_df = pd.read_csv('netflix_titles.csv')

netflix_df.tail (10)
```

```python
#show first 10 rows of data

netflix_df = pd.read_csv('netflix_titles.csv')

netflix_df.head (10)
```

```python
#present all necessary info of data:
#number of rows and columns
#number of non-null rows per column
#data type of each column

netflix_df.info()
```

<h3>DATA CLEANING</h3>

Given the question, only the year of when the movie/show was added to Netflix is required. So, split the date_added and drop the its first part, which includes the month and day.

```python
temp = netflix_df["date_added"].str.split(", ", n = 1, expand = True)
netflix_df["date_added"] = temp[1]

netflix_df.head(5)
```

Additionally, only the following columns are needed: title, country, date_added, release_year. The rest of the columns are to be discarded.

```python
columns = ['show_id', 'type', 'director', 'cast', 'rating', 'duration', 'listed_in', 'description']
netflix_df = netflix_df.drop(columns, axis=1)
netflix_df.info()
```

Only rows with a value for country that includes "South Korea" is needed.

```python
netflix_df = netflix_df[netflix_df['country'] == 'South Korea']
netflix_df.info()
```

The values for date_added also needs to be in the appropriate numerical data type in order to use them in mathematical operations.

```python
netflix_df['date_added'] = netflix_df['date_added'].astype(int)
netflix_df.dtypes
```

<h3>DATA ANALYSIS II</h3>

First, I want to know the number of South Korean movies and shows were added to Netflix per year.

```python
sk = netflix_df.date_added.value_counts(sort=False)
sk.name = "date_added_cnt"
sk_df = sk.to_frame()
sk_df.head(6)
```

The difference of the release year to the year it was added to Netflix is essential. So another column for it is added.

```python
netflix_df['difference_year'] = netflix_df['date_added']-netflix_df['release_year']
netflix_df.head(5)
```

Then, another dataframe is created to find the mean of the value of difference_year (representing the difference of the release year and the year it was addeed to Netflix) per release_year (representing the year it was released).

```python
diff = netflix_df.groupby(['release_year'])['difference_year'].mean()
diff.name = "diff_year_mean"
diff_df = diff.to_frame()
diff_df.tail(10)
```

<h3>DATA VISUALIZATION</h3>

First, visualize the sk_df dataframe.

As seen below, there has been an evident increase of the number of South Korean movies and shows added to Netflix through the past years. (The sudden decrease by the year of 2021 may be disregarded because the given data may not yet be newly updated and the year 2021 has not yet ended.) This may be caused by the heavy demand for such due to the steady rise of the Hallyu wave, or the global popularity of South Korean culture. The slight decrease in 2020 (compared to the past year of 2019) may be because of the unprecedented COVID-19 situation, which cancelled and delayed many projects globally.

```python
x = sk_df.index
y = sk_df.date_added_cnt

plt.style.use('seaborn-darkgrid')
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
plt.plot(x,y)

plt.title('Number of Movies Added vs Year')
plt.ylabel('Number of Movies Added' )
plt.xlabel('Year')
plt.show()
```

Second, visualize the diff_df dataframe.

Given that Netflix only started adding Korean shows and movies by 2016 (based on the previous table above), the number of relatively "old" (definition for this situation: before the current year) movies and shows added to Netflix has significantly decreased. All those initially released by 2019 onwards are immediately added to the content platform within the same year. This may be caused by the increase in the number of Korean shows and movies that are Netflix Originals.

```python
x = diff_df.index
y = diff_df.diff_year_mean
xi = list(range(len(x)))

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(111)
plt.plot(xi,y)
plt.xticks(xi,x)

plt.title('Mean of Difference vs Release Year')
plt.ylabel('Mean of the Difference of Release Year and Year Added' )
plt.xlabel('Release Year of Movie/Show')
plt.show()
```
